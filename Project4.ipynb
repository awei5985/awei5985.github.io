```python
import folium
import requests
import pandas
from folium.plugins import MarkerCluster
import os

arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")

arrest_table["race_new"] = arrest_table["sex"]
arrest_table["sex_new"] = arrest_table["race"]
arrest_table["race"] = arrest_table["race_new"]
arrest_table["sex"] = arrest_table["sex_new"]
arrest_table = arrest_table.drop('race_new', 1)
arrest_table = arrest_table.drop('sex_new', 1)

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

a_table = arrest_table.head(500)
a_table = a_table.reset_index(drop=True)
a_table
```

```python
locations = a_table[['lat','long']]
locationlist = locations.values.tolist()
length = len(locationlist)
```

```python
def gender(counter):
    if counter['sex'] == 'M':
        return 'male'
    else:
        return 'female'

def r_colors(counter):
    if counter['race'] == 'B':
        return 'saddlebrown'
    elif counter['race'] == 'W':
        return 'blue'
    elif counter['race'] == 'A':
        return 'tan'
    elif counter['race'] == 'H':
        return 'peru'
    elif counter['race'] == 'I':
        return 'orange'
    else:
        return 'black'
        

a_table["gender"] = a_table.apply(gender, axis=1)
a_table["r_colors"] = a_table.apply(r_colors, axis=1)
a_table.head()
```

```python
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)

marker_cluster = MarkerCluster().add_to(map_osm)

for point in range(0, length):
    folium.Marker(locationlist[point], popup = 'Race: '+a_table['race'][point]+', Location: '+a_table['arrestLocation'][point]+', Offense: '+a_table['incidentOffense'][point], 
                  icon = folium.Icon(color = a_table["r_colors"][point], icon_color = 'white', icon = a_table["gender"][point], 
                                     angle = 0, prefix = 'fa')).add_to(marker_cluster)
map_osm
```

```python
# So I wanted to plot a representation of what crime looks like in Boston based on the BPD_Arrest.csv. So I took the first 500
# entries and plotted them on the map. I plotted the first 500 entries because I opened up the actual csv file and looked at the
# arrest numbers. They didn't look they were organized in any particular order so I thought taking the first 500 entries would
# still result in a random sample so to speak. As for my map, I used marker clusters, because I felt that showing all 500 markers
# at the same time on the map could get messy and would be hard to read. Instead, there are clusters of markers that are
# clickable, and the map will zoom in to get a closer look. In addition, the markers let you see where crime occurs in clusters.
# 
# The number in the cluster represents the number of markers in said cluster, and hovering over a cluster will show you the area
# in which the markers are located. Each marker represents a crime. Each marker icon is color coded and has the male or female
# icon, representing the gender of the criminal. Each color represents a different race of criminal: blue for white, red brown
# for black, tan for asian, light brown for hispanic, orange for I (from the csv file. I don't know what race this is), and black
# if unknown. If you click the icon, the race, address, and criminal offense will pop up. The icons let you have a quick
# understanding of who committed a crime, and details are there if wanted.
```
