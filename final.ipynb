```python
# For my final project, I wanted to investigate whether or not franchise fatigue affects video game sales. Franchise fatigue is
# the concept where too many units of a franchise have been released to the point where people are bored of a franchise. As such
# sales go down. There is no formal definition, but the general idea is that as more and more entries of a certain franchise
# enters the market, the less the consumers are going to want to buy it, and thus, sales of said franchise reduce over time. This
# is most commonly applied to movies and video games. I want to see if franchise fatigue actually affects sales.
#
# Gaming websites and newsoutlets are always saying that fatigue has set into certain gaming franchises, and I want to see if
# it actually has.
#
# One such franchise I want to test is the Call of Duty franchise created by Infinity Ward, Treyarch, and Sledgehammer Games. 
# Call of Duty started in 2003 with Call of Duty, and has since spread into a massive gaming franchise covering, as of December
# 2017, 14 seperate games. These are, in order of release Call of Duty, Call of Duty 2, Call of Duty 3, 
# Call of Duty 4: Modern Warfare, Call of Duty: World at War, Call of Duty: Modern Warfare 2, Call of Duty: Black Ops
# Call of Duty: Modern Warfare 3, Call of Duty: Black Ops II, Call of Duty: Ghosts, Call of Duty: Advanced Warfare,
# Call of Duty: Black Ops III, Call of Duty: Infinite Warfare, and Call of Duty: WII.
# The data set I gathered only has data up to October 26, 2016, so I'm only going to be looking at the first 13 games released.
# In addition, the data set seems to not contain data for the first Call of Duty game, so that won't be included either.
#
# As a hypothesis, I do believe that franchise fatigue has set in in the Call of Duty franchise, meaning that people are getting
# bored of the yearly Call of Duty releases, which means sales have gone down. This will be evidenced by a downward trend in
# the game sales share percentage.
#
# List of major Call of Duty releases
# https://en.wikipedia.org/wiki/Call_of_Duty
#
# CoD Fatigue News articles
# https://www.forbes.com/sites/insertcoin/2013/11/06/call-of-duty-ghosts-is-modern-military-shooter-fatigue-incarnate/#705b51307686
# http://www.criticalhit.net/entertainment/call-duty-suffering-franchise-fatigue/
# 
# Dataset can be found here:
# https://www.kaggle.com/gregorut/videogamesales/data
```
